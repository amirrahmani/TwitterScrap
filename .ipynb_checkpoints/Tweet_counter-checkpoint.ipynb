# Analyzing the timming of the collected tweets.
## We assume the days are from noon to noon. For exampe Friday starts from previous Thursday noon and end at Friday noon. 
## Also we can change the start day of the week, for example we can study the number of tweets between each Friday recursively, or between any other day of the week.

```python
# load modules
import numpy as np
import pytz
import datetime 
from datetime import date, timedelta
import calendar
import sys
import tweepy
from tweepy import Stream
from tweepy import OAuthHandler
from tweepy.streaming import StreamListener
import matplotlib
matplotlib.use('tkAgg')
import matplotlib.pyplot as plt
import pendulum
import pandas as pd
%matplotlib inline
plt.rcParams['font.size']=20
plt.rcParams["font.family"] = "Times New Roman"
```

```python
sys.argv = ['','vp','Sat']  # the account user name and the first day of the week 

# Note that times in .txt file are in GMT 
tweet_file  =  sys.argv[1] + '_tweet_dates.txt' 
startday = sys.argv[2]
start_week_day = np.where(np.array(list(calendar.day_abbr))==startday)[0][0]

current_time = datetime.datetime.now()

# authenticate twitter
API_KEY='' #25 alphanumeric characters
API_SECRET='' #50 alphanumeric characters
ACCESS_TOKEN = ''
ACCESS_TOKEN_SECRET = ''

# This is not necessary
# class TweetListener(StreamListener):
#     # A listener handles tweets received from the stream.
#     # This is a basic listener that just prints received tweets to standard output
#     def on_data(self, data):
#         print(data)
#         return True
#     def on_error(self, status):
#         print(status)


# auth = OAuthHandler(API_KEY,API_SECRET)
# api = tweepy.API(auth)

# auth.set_access_token(ACCESS_TOKEN, ACCESS_TOKEN_SECRET)
# twitterStream = Stream(auth,TweetListener())
# user = api.get_user(sys.argv[1])

utc=pytz.utc
eastern=pytz.timezone('US/Eastern')

counterStartYear = 2017
counterStartMonth = 'Oct'
counterStartDate = 1

start_day_count = datetime.datetime.strptime(str(counterStartDate)+' '\
                                             +counterStartMonth+' '+\
                                             str(counterStartYear),\
                                             '%d %b %Y')
```

```python
# read the file and make the dataframe
df = pd.read_table(tweet_file,names=['date'])
df['datetime'] = df['date'].apply(lambda x:datetime.datetime.strptime\
                                  (x.strip(),"%a %b %d %H:%M:%S +0000 %Y"))
#df['datetime'] =pd.to_datetime(df.date)  # another way to do the above

# changing the time zone to Eastern time zone (NYC)
df['utc'] = df['datetime'].apply(lambda x: utc.localize(x,is_dst=None))
df['eastern_noshift'] = df['utc'].apply(lambda x: x.astimezone(eastern))
# shifting the tweets in the afternoon to the next day
df['eastern'] = df['eastern_noshift'].apply(lambda x: \
                                            x+datetime.timedelta(days=1)\
                                            if x.hour>=12 else x)
df.drop(['utc'],axis = 1)
    

Tmaxid = df['eastern'].diff().idxmax() 
Tdiff = df['eastern'].diff().max        # maximum time between two tweets
# create a column for dates in the dd mm YYYY format
df['idx'] = df['eastern'].apply(lambda x:x.strftime("%d")+\
                                ' '+x.strftime("%b")+' '+x.strftime("%Y"))

df['morning'] = df['eastern'].apply(lambda x: x.hour<12)  # tweets in the morning
```

```python
x = df.groupby(['idx']).agg(['count'])['date']         # count number of tweets for each date
x.index = pd.to_datetime(x.index)
x.sort_index(level=0,inplace=True)
# add another index with day name
x.index = pd.MultiIndex.from_tuples([(z, str(z.strftime("%a")))\
                                      for z in x.index])
x.index.rename(['date','day'],inplace=True)

y = df.groupby(['idx','morning']).agg(['count'])['date']    # same as x but with separating morning tweets
y.index = pd.MultiIndex.from_tuples([(pd.to_datetime(z[0]),\
                                      pd.to_datetime(z[0]).strftime("%a"),\
                                       z[1]) for z in y.index])
y.index.rename(['date','day','ismorning'],inplace=True)
y.sort_index(level=0,inplace=True)
```

```python
def weekly_day(dname,y,ismorning=True):
    if ismorning:
        z = y.loc[idx[:,dname,True],idx[:]][-10:]
    else:
        z= y.loc[idx[:,[dname]],idx[:]][-10:]
    z.reset_index(inplace=True)
    start = z.date.iloc[-1]
    dates = []
    while start>=z.date.iloc[0]:
        dates.append(start)
        start += timedelta(-7)
    dates = pd.DataFrame(dates,columns=['date'])
    z = pd.merge(z,dates,how='right',on=['date'])
    z.sort_values(['date'],inplace=True)
    z.fillna(0,inplace = True)
    z['count'] = z['count'].apply(lambda x: int(x))
    return(z['count'][-10:].values)



# counting the number of tweets for each day of week
idx = pd.IndexSlice
allday_count = {}
morning_count = {}
for dname in list(calendar.day_abbr):
    allday_count[dname] = weekly_day(dname,x,False)
    morning_count[dname] = weekly_day(dname,y)
#    morning_count[dname] = y.loc[idx[:,dname,True],idx[:]][-10:]['count'].values


# defining the equivalant of today in previous weeks
current_end_week = current_time
if current_end_week.weekday() == start_week_day-1 and \
   current_end_week.hour >= 12:
    current_end_week += timedelta(weeks=1)
else:
    while current_end_week.weekday() != start_week_day-1:
        current_end_week += timedelta(days=1)

current_end_week = current_end_week.replace(hour=12,minute=00,second=00)
current_end_week2 = current_end_week
current_start_week = current_end_week + timedelta(weeks=-1)
current_start_week2 = current_end_week + timedelta(weeks=-1)
ttend=[]
ttweek=[]
ntweet=[]
htweet=[]
thour = []
nhour = []
tweek=0
tend=0
current_time_in_week = current_time
ff = current_time-current_start_week

# counting number of tweets for each week
for counter,d in enumerate(reversed(df['eastern'])):
    if d.hour>=12:
        dummyd = d - timedelta(days=1)
    else:
        dummyd = d
    dummyd = dummyd.replace(tzinfo=None)
    if dummyd<current_time_in_week:
        current_time_in_week += timedelta(weeks=-1) 
        current_end_week += timedelta(weeks=-1)
        ttend.append(tend)
        tend = 0

    if dummyd<current_start_week2:
        current_start_week2 += timedelta(weeks=-1) 
        current_end_week2 += timedelta(weeks=-1)
        ttweek.append(tweek)
        ntweet.append(range(0,len(thour)))
        htweet.append(list(reversed(thour)))
        thour = []
        nhour = []
        tweek = 0

    if dummyd<current_end_week2 and dummyd>current_start_week2:
        tweek += 1  
        xhour = dummyd-current_start_week2
        xhour = (xhour.days*24*3600+xhour.seconds)/3600.
        thour.append(xhour)
        nhour.append(tweek)

    if dummyd>current_time_in_week and dummyd<current_end_week:
        tend += 1
```

## Print and plot results

```python
print('Number of tweets for each day at each week (the most recent week is the one on the right):\n')
print('                All Day            ','         Morning         ')
for d in list(calendar.day_abbr):
    print(d,end="")
    for x in allday_count[d]:
        print('%3d' % x,end="")
    print(' | ',end="")    
    for x in morning_count[d]:
        print('%3d' % x,end="")
    print(' ')

print('\nNumber of tweets at each week (the latest week is the one at the left)')
print('week count: ', ttweek)

print('\nNumber of tweets from the same current day in the previos week to the end of the week (the latest week is the one at the left)')
print('tweet to end: ', ttend)

ymax = 0
handle1 = []
fig1 = plt.figure(1)
ax1 = fig1.add_subplot(111)


for i in range(6):
      handle1.append(ax1.plot(htweet[i],ntweet[i],'-o',\
                              label = 'week ' + str(i)))
      ymax = max(ymax,ntweet[i][-1])

legend = plt.legend(loc='upper left',frameon=False)

for i in range(1,8):
      ax1.axvline(i*24,color='k',linestyle='--')

ax1.axvline(ff.days*24+ff.seconds/3600.,color='r')
ax1.set_xlim(0,7*24)
ax1.set_xlabel('hours')
ax1.set_ylabel('n')
ax1.grid(True)
fig1.set_size_inches(12,6)
```
